## Question

`1) Find the maximum riders on the First day`

`2) Find the mean rider per day` 

## 2D Array

```python
def mean_raider_max_station(ridership):
    # Find the station with maximum riders on the first day
    max_station = ridership[0, :].argmax()

    #Find the mean rider per day for that station
    mean_for_max = ridership[:, max_station].mean()

    #Find the mean ridership overall for comparions
    overall_mean = ridership.mean()

    return (overall_mean, mean_for_max)
```

```python
mean_raider_max_station(ridership)
```

## Question

`3) Find the mean rider per day for each subway station`

`4)Find the Min and Max daily Raidership per day `

```python
def mean_and_max_raider_per_day(ridership):
    #Find the mean raider per day of each subway station
    station_raiders = ridership.mean(axis=0)
    
    # Find the Min and Max daily Raidership
    max_daily_raidership = station_raiders.max()
    min_daily_raidership = station_raiders.min()
    
    return (max_daily_raidership, min_daily_raidership)
```

```python
mean_and_max_raider_per_day(ridership)
```

## Finding correlation 

```python
def correlation(x, y):
    std_x = (x-x.mean()) / x.std(ddof=0)
    std_y = (y-y.mean()) / y.std(ddof=0)
    
    return (std_x * std_y).mean()
```

```python
entries = subway_df['ENTRIESn_hourly']
cum_entries = subway_df['ENTRIESn']
rain = subway_df['meanprecipi']
temp = subway_df['meantempi']

print(correlation(entries, rain))
print(correlation(entries, temp))
print(correlation(rain, temp))
print(correlation(entries, cum_entries))
```

## Convert numerical grades to letter grades 
 
    90 - 100 = A
    80 - 89  = B
    70 - 79  = c
    60 - 69  = D
    0  - 59  = F 

```python
grades_df = pd.DataFrame(
    data={'exam1': [43, 81, 78, 75, 89, 70, 91, 65, 98, 87],
          'exam2': [24, 63, 56, 56, 67, 51, 79, 46, 72, 60]},
    index=['Andre', 'Barry', 'Chris', 'Dan', 'Emilio', 
           'Fred', 'Greta', 'Humbert', 'Ivan', 'James']
)
```

```python
def convert_grade(grade):
    if grade >= 90:
        return 'A'
    elif grade >= 80:
        return 'B'
    elif grade >= 70:
        return 'C'
    elif grade >= 60:
        return 'D'
    else:
        return 'F'
```

```python
def convert_grades(grades):
    return grades.applymap(convert_grade)
```

```python
convert_grades(grades_df)
```

## Fill in this function to return the second-largest value of each column of the input DataFrame.

```python
def second_largest_in_column(column):
    sorted_column = column.sort_values(ascending=False)
    return sorted_column.iloc[1]
```

```python
second_largest_in_column(df['a'])
```

```python
def second_largest(df):
    return df.apply(second_largest_in_column)
```
